```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("iris.csv")
```

**tail()** ->

```python
#Plots a scatter plot for length and width of all Petal for all species
plt.scatter(data["petal_length"],data["petal_width"], c=data["species"].map({"setosa" :"orange", "versicolor" : "blue", "virginica" : "yellow"}))
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.legend()
plt.title("Petal length vs Petal width")
plt.show()
```

```python
#Plots a scatter plot for length and width of all sepal for all species
plt.figure(figsize=(8,6))
sns.scatterplot(data=data, x="sepal_length", y="sepal_width", hue="species")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Sepal Width")
```

```python
#Creates a box plot for sepal width with outliers
ax_sw = sns.boxplot(x=data["species"], y=data["sepal_width"], hue=data["species"])
```

```python
#Creates a box plot for sepal length with outliers
ax_sl = sns.boxplot(x=data["species"], y=data["sepal_length"], hue=data["species"])
```

```python
#Creates a box plot for petal width with outliers
ax_pw = sns.boxplot(x=data["species"], y=data["petal_width"], hue=data["species"])
```

```python
#Creates a box plot for petal length with outliers
ax_pl = sns.boxplot(x=data["species"], y=data["petal_length"], hue=data["species"])
```

```python
#Create a boxplot to detect outliers of whole data
sns.boxplot(data=data)
plt.show()
```

```python
#Creates a histogram
data.hist(figsize=(10, 8))
plt.show()
```
